# RAG

```python
pdf_path = r"/Users/saro/Downloads/desktop june 2026/langchain/rag_project/Saravana_Kumar_T_Resume.pdf"
from langchain_community.document_loaders import PyPDFLoader
loader = PyPDFLoader(pdf_path)
documents = loader.load()
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=200
)

docs = splitter.split_documents(documents)

print(docs)
print(len(docs))
```

# Embedding

```python
from  langchain_google_genai import ChatGoogleGenerativeAI, GoogleGenerativeAIEmbeddings
from langchain_community.vectorstores import FAISS
```

```python
embeddings = GoogleGenerativeAIEmbeddings(
    model="models/gemini-embedding-2"
)
```

```python
FAISS.from_documents?
```

```python
vector_store = FAISS.from_documents(
    docs,embeddings
)
```

```python
dir(vector_store)
```

```python
vector_store.as_retriever?
```

```python
retriever = vector_store.as_retriever(
    search_kwargs={"k":3}
)
```

```python
from langchain_core.prompts import PromptTemplate
prompt = PromptTemplate(
    input_variables=["context","question"],
    template = """
You are an intelligent assistant

use only the provided context to answer

context:
{context}

question:
{question}

Answer:
    """
)
```

```python
while True:

    question = input("\nAsk a Question (type exit to quit): ")

    if question.lower() == "exit":
        break

    # Retrieve relevant chunks
    retrieved_docs = retriever.invoke(question)

    context = "\n\n".join(
        [doc.page_content for doc in retrieved_docs]
    )

    final_prompt = prompt.format(
        context=context,
        question=question
    )

    response = llm.invoke(final_prompt)

    print("\nAnswer:")
    print(response.content)

    print("\n" + "=" * 80)
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.document_loaders import PyPDFLoader


pdf_path = r"/Users/saro/Downloads/desktop june 2026/langchain/rag_project/Saravana_Kumar_T_Resume.pdf"

loader = PyPDFLoader(
    pdf_path
)

documents = loader.load()
print(documents)

splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap = 100
)

chunking = splitter.split_documents(documents)

print(chunking)
```

```python
from langchain_google_genai import GoogleGenerativeAIEmbeddings, ChatGoogleGenerativeAI

embeddings = GoogleGenerativeAIEmbeddings(
    model = "models/gemini-embedding-2"
)
```

```python
from langchain_community.vectorstores import FAISS
```

```python
help(FAISS)
```

```python
vectorstore = FAISS.from_documents(
    chunking,
    embeddings
)
```

```python
help(vectorstore)
```

```python
retriever = vectorstore.as_retriever(
    search_kwargs={"k":3}
)
```

```python
llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",
    temperature=0
)
```

```python
from langchain_core.prompts import PromptTemplate

prompt = PromptTemplate(
    input_variables=['context','question'],
    template="""
    You are an excellent AI assistant,

    You should answer well grounded to the context

    context:
    {context}

    question:
    {question}

    Answer:
    """
)
```

```python
while True:
    question = input("Enter your question or type exit to close")

    if question.lower() == 'exit':
        break

    retrieved_docs = retriever.invoke(question)

    context = "\n\n".join(
        [doc.page_content for doc in retrieved_docs]
    )

    final_prompt = prompt.format(
        context=context,
        question=question
    )

    response = llm.invoke(question)

    print(response)
```
