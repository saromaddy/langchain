```python
from langchain_community.document_loaders import PyPDFLoader
pdf_path = "Saravana_Kumar_T_Resume.pdf"

loader = PyPDFLoader(pdf_path)
documents = loader.load()
```

```python
print(documents[1])
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=300
)

docs = splitter.split_documents(documents)

print(f"Created {len(docs)} chunks")
```

```python
docs[0].page_content
```

```python
docs[1].page_content
```

```python
from langchain_community.vectorstores import FAISS
from langchain_google_genai import GoogleGenerativeAIEmbeddings
embeddings = GoogleGenerativeAIEmbeddings(
    model="models/gemini-embedding-2"
)

# --------------------------------------
# STEP 4: Create Vector Database
# --------------------------------------

vector_store = FAISS.from_documents(
    docs,
    embeddings
)
```

```python
print(vector_store)
```

```python
import pandas as pd
from IPython.display import display

# --------------------------------------
# STEP 4: Create Vector Database (Your existing code)
# --------------------------------------
# vector_store = FAISS.from_documents(docs, embeddings)

# --------------------------------------
# STEP 5: Visualize the Vector Store in Jupyter
# --------------------------------------

# LangChain's FAISS implementation stores the original text chunks and metadata 
# inside the 'docstore' attribute. We access its hidden dictionary (_dict).
docstore_data = vector_store.docstore._dict

# Initialize an empty list to gather the extracted data
extracted_rows = []

# Iterate through the docstore dictionary to extract the ID, content, and metadata
for doc_id, doc_object in docstore_data.items():
    extracted_rows.append({
        "Document ID": doc_id,
        "Page Content": doc_object.page_content,
        "Metadata": doc_object.metadata
    })

# Convert the list of dictionaries into a Pandas DataFrame for tabular visualization
df_vector_store = pd.DataFrame(extracted_rows)

# The display() function renders the DataFrame as a formatted, interactive table in ipynb
display(df_vector_store)

# --------------------------------------
# OPTIONAL: Viewing the actual Embedding Vectors (Numbers)
# --------------------------------------
# If you want to see the raw array of numerical embeddings (the vectors themselves),
# you can access the underlying FAISS index.

# Get the total number of vectors stored
total_vectors = vector_store.index.ntotal
print(f"\nTotal vectors in the index: {total_vectors}")

# Reconstruct and print the raw numerical vector for the first item (Index 0)
# Note: This will be a very long array of floating-point numbers
if total_vectors > 0:
    first_vector = vector_store.index.reconstruct(0)
    print("\nRaw numerical embedding for the first chunk (truncated):")
    print(first_vector[:10], "... [array continues]")
```
